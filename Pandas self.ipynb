```python
# create series using dictionary.
#index values is change which is index values as a key
dict={101:"Jivan",102:"pravin",103:"pratik",104:"Shubham"}
se=pd.Series(dict)
print(se)
```

```python
#index change in list as tuple
Ls=["Monday","Tuesday","Wednesday","Thresday","Friday"]
Index=["day1", "day2", "day3", "day4", "day5"]
se=pd.Series(Ls,(Index))
print(se)

#OR

Ls = pd.Series(["Monday","Tuesday","Wednesday","Thursday","Friday"])
Ls.index = [101,102,103,104,105]
print("\n second method :")
print(Ls)
```

## Combining DataFrames
- All arrays must be of the same length

### Row Cmobin

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3','A4','A5'],
                  'B':['B0','B1','B2','B3','B4','B5'],
                  'C':['C0','C1','C2','C3','C4','C5'],
                  'D':['D0','D1','D2','D3','D4','D5']})
df1
```

```python
df2=pd.DataFrame({'A':['A0','A1','A2','A3','A4','A5'],
                  'B':['B0','B1','B2','B3','B4','B5'],
                  'C':['C0','C1','C2','C3','C4','C5'],
                  'D':['D0','D1','D2','D3','D4','D5']})
print(df2)
```

```python
# Row marging/concatinate
pd.concat([df1,df2],axis=0,ignore_index=True)
```

### for Column

```python
# Column concatinate/combine of two frame.(axis=1)for column
# ignore index is requred for naming of columns.ignore_index=False
pd.concat([df1,df2],axis=1)
```
