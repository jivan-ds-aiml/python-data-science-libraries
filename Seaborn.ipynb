```python
# import basic libraryes
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
#seaborn has inbuilt dataset 
sns.get_dataset_names()
```

```python
#import iris dataset.
data=sns.load_dataset('iris')
data
```

## Relational plots
 relational plot are used to understand relationship between two variables
- line plot
- scatter plot
- relplot

```python
# Line plot
#Each point/line position shows the relationship between sepal length and sepal width.
#If sepal_length increases and sepal_width also increases, they have a positive relationship.
#If sepal_length increases but sepal_width decreases, they have a negative relationship.
sns.lineplot(data=data,x='sepal_length',y='sepal_width',color='red')
```

```python
#For the Iris dataset, lineplot is not the best choice because the data is not time-series.(use scatterplot)
#(ci)Confidance interval, hue use for categorical columns
sns.lineplot(data=data,x='sepal_length',y='sepal_width',hue='species',ci=None)
```

```python
## Scatter plot
```

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length')
plt.show()
# here positive relation in two variable
```

```python
sns.scatterplot(data=data,x='sepal_length',y='sepal_width')
plt.show()
```

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')

#Setosa → slight relationship
#Versicolor → weak relationship
#Virginica → weak relationship
```

```python
sns.scatterplot(data=data, x='sepal_length', y='petal_length', hue='species',palette='rainbow')
```

## rel plot

```python
# relplot by default it is scatter plot
#relplot can be converted into line plot by giving kind='line'
sns.relplot(data=data,x='petal_length',y='sepal_width',kind='line',ci=None)
# no relation
```

## Categorical plots
it is used to analyse one numerical and one categorical column.
- count plot
- box plot
- barplot
- violin plot
- strip Plot
- swarm plot
- cat plt

## count plot
- it hust gives count of categories

```python
sns.countplot(data=data,x='species',palette='rainbow')
# equal record- Balance data
```

```python
# Bar plot
#compare one categoricalcolumn and one is numerical column here.
sns.barplot(data=data,x='species',y='sepal_length',palette='Set1')
plt.show()
```

```python
sns.barplot(data=data,x='species',y='petal_width',palette='Set1')
plt.show()
```

# Box Plot
Box plot is used to identify outliers


```python
sns.boxplot(data,x='sepal_length')
plt.show()
```

```python
sns.boxplot(data=data,x='species',y='sepal_length',palette='Set1')
#palette='Set1'(use for coloring)
# virginica has higher sepal_length
# versicolor has high middle
```

## Violin plot

```python
sns.violinplot(data,x='sepal_width')
plt.show()
# it is like a box plot.interpretation is same.
```

```python
sns.violinplot(data=data,x='species',y='sepal_width',palette='Set1')
# setosa has higher sepal_width
# versicolor has min sepal_with
```

## Strip Plot

```python
# it overlap the plot
sns.stripplot(data,x='species',y='sepal_width',hue='species')
plt.show()
```

```python
# swarmplot :- it dose not overlap the plot
sns.swarmplot(data,x='species',y='sepal_width',hue='species')
```

# cat plot 
using cat plot we can plot box,violin,strip,swarm.

```python
sns.catplot(data=data,x='species',y='sepal_length',hue='species')
# it by default stripplot
sns.catplot(data=data,x='species',y='sepal_length',hue='species',kind='box')
```

```python
sns.catplot(data=data,x='species',y='sepal_length',hue='species',kind='violin')
```

## Distribution Plot
- histplot
-  dist plot
* this two graphs are used to check distribution of data
* use for only numerical values

```python
sns.histplot(data=data,x='petal_width',kde=True,hue='species')
```

```python
#distplot
# it show bell shape of data with histogram.
sns.distplot(x=data['sepal_length'])
```

```python
# pareplot
sns.pairplot(data=data,hue='species')
```
